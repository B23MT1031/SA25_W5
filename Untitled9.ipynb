Feature engineering

```python
def make_features(df):
    df = df.copy()
    df['occ_rate']   = df['occupancy'] / df['capacity']
    df['queue_norm'] = df['queue_length'] / df.groupby('lot_id')['queue_length'].transform('max').clip(lower=1)
    df['traffic_z']  = (df['traffic'] - df['traffic'].mean()) / df['traffic'].std()

    vmap = {'car': 0, 'bike': -0.1, 'truck': 0.25}
    df['veh_weight'] = df['vehicle_type'].map(vmap)

    return df
```

model 1

```python
def smooth_price(p_prev, p_raw, max_step=1.00):
#Limit change to ±$1.00 per 30‑min slot so curves look nice.
    return np.clip(p_raw, p_prev - max_step, p_prev + max_step)
```

```python
ALPHA = 8.0
BASE  = 10.0
BASE_PRICE = 10.0


def price_m1(row, p_prev):
    target = BASE + ALPHA * row['occ_rate']
    return smooth_price(p_prev, target)
```

model 2

```python
def make_features(df):
    df = df.copy()
    df['occ_rate'] = df['Occupancy'] / df['Capacity']
    df['queue_norm'] = df['QueueLength'] / df.groupby('ID')['QueueLength'].transform('max').clip(lower=1)
    df['traffic_z'] = (df['TrafficConditionNearby'] - df['TrafficConditionNearby'].mean()) / df['TrafficConditionNearby'].std()

    vmap = {'car': 0, 'bike': -0.1, 'truck': 0.25}
    df['veh_weight'] = df['VehicleType'].map(vmap)

    return df

```

```python
# Convert categorical traffic to numeric levels
traffic_map = {'low': 1, 'medium': 2, 'high': 3}
df['TrafficConditionNearby'] = df['TrafficConditionNearby'].map(traffic_map)

# Convert IsSpecialDay to 0/1 if it's not already numeric
df['IsSpecialDay'] = df['IsSpecialDay'].astype(int)

# Convert QueueLength, Capacity, Occupancy to numeric (forcefully if needed)
df['QueueLength'] = pd.to_numeric(df['QueueLength'], errors='coerce')
df['Capacity'] = pd.to_numeric(df['Capacity'], errors='coerce')
df['Occupancy'] = pd.to_numeric(df['Occupancy'], errors='coerce')
```

```python
df = make_features(df)
```

```python
def raw_demand(row):
    a, b, c, d, e = 1.2, 0.8, 0.5, 1.4, 1.0
    return (a * row['occ_rate'] +
            b * row['queue_norm'] -
            c * row['traffic_z'] +
            d * row['IsSpecialDay'] +
            e * row['veh_weight'])

df['demand'] = df.apply(raw_demand, axis=1)
D_MEAN = df['demand'].mean()
D_STD = df['demand'].std()
```

Model 2 pricing

```python
λ = 0.8
bounds = (0.5, 2)  # Price range: $5–$20

def price_m2(row, p_prev):
    a, b, c, d, e = 1.2, 0.8, 0.5, 1.4, 1.0

    demand = (a * row['occ_rate'] +
              b * row['queue_norm'] -
              c * row['traffic_z'] +
              d * row['IsSpecialDay'] +
              e * row['veh_weight'])

    norm_d = (demand - D_MEAN) / D_STD
    raw_price = BASE_PRICE * (1 + λ * norm_d)
    raw_price = np.clip(raw_price, BASE_PRICE * bounds[0], BASE_PRICE * bounds[1])
    return smooth_price(p_prev, raw_price)
```

Example

```python
price_state = {}
output = []

for _, row in df.iterrows():
    lot = row['ID']
    prev_price = price_state.get(lot, BASE_PRICE)

    new_price = price_m2(row, prev_price)  # or price_m1(row, prev_price)
    price_state[lot] = new_price

    output.append({
        'timestamp': row['timestamp'],
        'lot_id': lot,
        'price': new_price
    })

price_df = pd.DataFrame(output)
```

compare model 1 & model 2


```python
df = make_features(df)  # make sure to run this first

price_state_m1 = {}
price_state_m2 = {}
outputs = []

for _, row in df.iterrows():
    lot = row['ID']

    p1_prev = price_state_m1.get(lot, BASE_PRICE)
    p2_prev = price_state_m2.get(lot, BASE_PRICE)

    p1 = price_m1(row, p1_prev)
    p2 = price_m2(row, p2_prev)

    price_state_m1[lot] = p1
    price_state_m2[lot] = p2

    outputs.append({
        'timestamp': row['timestamp'],
        'lot_id': lot,
        'price_m1': p1,
        'price_m2': p2
    })

compare_df = pd.DataFrame(outputs)
compare_df['timestamp'] = pd.to_datetime(compare_df['timestamp'])
```

```python
import matplotlib.pyplot as plt

lots_to_plot = compare_df['lot_id'].unique()[:3]  # first 3 lots

fig, axes = plt.subplots(nrows=len(lots_to_plot), ncols=1, figsize=(14, 4 * len(lots_to_plot)))

if len(lots_to_plot) == 1:
    axes = [axes]  # ensure iterable

for ax, lot_id in zip(axes, lots_to_plot):
    lot_df = compare_df[compare_df['lot_id'] == lot_id]

    ax.plot(lot_df['timestamp'], lot_df['price_m1'], label='Model 1 (Baseline)', color='blue')
    ax.plot(lot_df['timestamp'], lot_df['price_m2'], label='Model 2 (Demand-Based)', color='green')

    ax.set_title(f"Lot {lot_id} — Price Over Time")
    ax.set_xlabel("Time")
    ax.set_ylabel("Price ($)")
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
